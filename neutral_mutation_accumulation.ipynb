```python
from pathlib import Path
import pickle as pkl
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = Path("results/simulation")
g = 1
experiment_name = f"model_N_1_g_{g}_beta_1"

z_nc = np.zeros((10, 1000), dtype=int)

for replica in range(1, 11):
    for generation in range(1, 1001):
        with open(path / experiment_name / str(replica) / f"logs/generation_{generation}.pkl", "rb") as pkl_file:
            stats = pkl.load(pkl_file)
        z_nc[replica - 1, generation - 1] = stats["population"]["z_nc_array"][0]

plt.plot([x for x in range(1, 1001)], z_nc.mean(axis=0))
plt.title("Mean evolution of non conding size (10 replicas)")
plt.xlabel("generation")
plt.ylabel("z_nc")
plt.plot()
        
```
